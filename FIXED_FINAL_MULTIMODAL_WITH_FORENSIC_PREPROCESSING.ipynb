```python
# =========================================================
# FINE-TUNED MULTIMODAL CLIP
# FORENSIC AUTOCORRELATION VERSION
# VISUAL ENCODER FINE-TUNING
# =========================================================

!pip install -q transformers datasets torchvision scikit-learn opencv-python seaborn

# =========================================================
# IMPORTS
# =========================================================

import os
import random
import numpy as np
import cv2

from PIL import Image

import torch
import torch.nn as nn

from torch.utils.data import Dataset, DataLoader

from datasets import load_dataset

from transformers import (
    CLIPProcessor,
    CLIPModel
)

from torchvision import transforms

from sklearn.preprocessing import label_binarize

from sklearn.metrics import (

    accuracy_score,

    f1_score,

    classification_report,

    confusion_matrix,

    precision_score,

    recall_score,

    roc_curve,

    auc,

    precision_recall_curve,

    average_precision_score
)

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from google.colab import drive

# =========================================================
# GOOGLE DRIVE
# =========================================================

drive.mount('/content/drive')

# =========================================================
# GLOBAL SEED
# =========================================================

SEED = 42

random.seed(SEED)

np.random.seed(SEED)

torch.manual_seed(SEED)

torch.cuda.manual_seed(SEED)

torch.cuda.manual_seed_all(SEED)

os.environ["PYTHONHASHSEED"] = str(SEED)

torch.backends.cudnn.deterministic = True

torch.backends.cudnn.benchmark = False

# =========================================================
# DEVICE
# =========================================================

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("🚀 Device:", device)

# =========================================================
# LOAD DATASET
# =========================================================

dataset = load_dataset(
    "Rajarshi-Roy-research/Defactify_Image_Dataset"
)

train_ds = dataset["train"]

val_ds = dataset["validation"]

test_ds = dataset["test"]

# =========================================================
# CLIP PROCESSOR
# =========================================================

processor = CLIPProcessor.from_pretrained(
    "openai/clip-vit-base-patch32"
)

# =========================================================
# FORENSIC PREPROCESSING
# =========================================================

def frequency_preprocessing(pil_image):

    image = np.array(
        pil_image.convert("RGB")
    )
# =========================================================
# RGB → GRAYSCALE
# =========================================================

    gray = cv2.cvtColor(
        image,
        cv2.COLOR_RGB2GRAY
    ).astype(np.float32) / 255.0
# =========================================================
# WEAK DENOISING
# =========================================================
    denoised = cv2.GaussianBlur(
        gray,
        (3, 3),
        0
    )
# =========================================================
# RESIDUAL EXTRACTION
# r = x - D(x)
# =========================================================
    residual = gray - denoised
# =========================================================
# FFT
# =========================================================
    fft = np.fft.fft2(
        residual
    )

    fft_shift = np.fft.fftshift(
        fft
    )
# =========================================================
# POWER SPECTRUM
# =========================================================
    power = np.abs(
        fft_shift
    ) ** 2
# =========================================================
# AUTOCORRELATION
# =========================================================
    autocorr = np.fft.ifft2(
        power
    )

    autocorr = np.real(
        autocorr
    )

    autocorr = np.fft.fftshift(
        autocorr
    )
# =========================================================
# LOG SCALE
# =========================================================
    autocorr = np.log1p(
        np.abs(autocorr)
    )
# =========================================================
# NORMALIZATION
# =========================================================
    autocorr = (
        autocorr - autocorr.min()
    ) / (
        autocorr.max() - autocorr.min() + 1e-8
    )
# =========================================================
# UNIT8
# =========================================================
    autocorr = (
       autocorr * 255
    ).astype(np.uint8)
# =========================================================
# GRAYSCALE → RGB
# =========================================================
    autocorr_rgb = cv2.cvtColor(
        autocorr,
        cv2.COLOR_GRAY2RGB
    )
# =========================================================
# PIL IMAGE
# =========================================================
    return Image.fromarray(
        autocorr_rgb
    )

# =========================================================
# AUGMENTATION
# =========================================================

train_transform = transforms.Compose([

    transforms.RandomResizedCrop(
        224,
        scale=(0.75, 1.0)
    ),

    transforms.RandomHorizontalFlip(),

    transforms.RandomRotation(10)

])

# =========================================================
# DATASET
# =========================================================

class MultiModalDataset(Dataset):

    def __init__(self, hf_dataset, train=False):

        self.dataset = hf_dataset

        self.train = train

    def __len__(self):

        return len(self.dataset)

    def __getitem__(self, idx):

        item = self.dataset[idx]

        image = item["Image"].convert("RGB")

        image = frequency_preprocessing(
            image
        )

        if self.train:

            image = train_transform(
                image
            )

        caption = item["Caption"]

        encoding = processor(

            text=caption,

            images=image,

            padding="max_length",

            truncation=True,

            max_length=77,

            return_tensors="pt"
        )

        return {

            "pixel_values":
                encoding["pixel_values"].squeeze(0),

            "input_ids":
                encoding["input_ids"].squeeze(0),

            "attention_mask":
                encoding["attention_mask"].squeeze(0),

            "label_B":
                torch.tensor(
                    item["Label_B"],
                    dtype=torch.long
                )
        }

# =========================================================
# DATALOADER SEEDING
# =========================================================

def seed_worker(worker_id):

    worker_seed = SEED + worker_id

    np.random.seed(worker_seed)

    random.seed(worker_seed)

g = torch.Generator()

g.manual_seed(SEED)

# =========================================================
# DATALOADERS
# =========================================================

BATCH_SIZE = 32

train_loader = DataLoader(
    MultiModalDataset(train_ds, train=True),
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

val_loader = DataLoader(
    MultiModalDataset(val_ds),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

test_loader = DataLoader(
    MultiModalDataset(test_ds),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

# =========================================================
# MODEL
# =========================================================

class FineTunedMultiModalCLIP(nn.Module):

    def __init__(self):

        super().__init__()

        self.clip = CLIPModel.from_pretrained(
            "openai/clip-vit-base-patch32"
        )

        for param in self.clip.text_model.parameters():

            param.requires_grad = False

        hidden_dim = self.clip.config.projection_dim

        fusion_dim = hidden_dim * 2

        self.fusion = nn.Sequential(

            nn.Linear(fusion_dim, 256),

            nn.LayerNorm(256),

            nn.GELU(),

            nn.Dropout(0.5),

            nn.Linear(256, 128),

            nn.LayerNorm(128),

            nn.GELU(),

            nn.Dropout(0.5)
        )

        self.classifier = nn.Linear(
            128,
            6
        )

    def forward(

        self,

        pixel_values,

        input_ids,

        attention_mask
    ):

        outputs = self.clip(

            pixel_values=pixel_values,

            input_ids=input_ids,

            attention_mask=attention_mask
        )

        image_feat = outputs.image_embeds

        text_feat = outputs.text_embeds

        fusion = torch.cat([

            image_feat,

            text_feat

        ], dim=1)

        fused = self.fusion(
            fusion
        )

        logits = self.classifier(
            fused
        )

        return logits

# =========================================================
# MODEL
# =========================================================

model = FineTunedMultiModalCLIP().to(
    device
)

# =========================================================
# LOSS
# =========================================================

criterion = nn.CrossEntropyLoss(
    label_smoothing=0.1
)

# =========================================================
# OPTIMIZER
# =========================================================

optimizer = torch.optim.AdamW(

    model.parameters(),

    lr=1e-6,

    weight_decay=1e-3
)

# =========================================================
# HISTORY
# =========================================================

train_losses = []
train_accuracies = []

val_losses = []

val_accuracies = []

val_f1s = []

# =========================================================
# TRAINING
# =========================================================

best_score = 0

EPOCHS = 25

for epoch in range(EPOCHS):

    print(f"\n🔥 Epoch {epoch+1}/{EPOCHS}")

    model.train()

    total_loss = 0
    train_preds = []
    train_targets = []

    for batch in tqdm(train_loader):

        pixel_values = batch["pixel_values"].to(device)

        input_ids = batch["input_ids"].to(device)

        attention_mask = batch["attention_mask"].to(device)

        labels = batch["label_B"].to(device)

        optimizer.zero_grad()

        outputs = model(

            pixel_values,

            input_ids,

            attention_mask
        )

        loss = criterion(
            outputs,
            labels
        )

        loss.backward()

        torch.nn.utils.clip_grad_norm_(
            model.parameters(),
            1.0
        )

        optimizer.step()

        total_loss += loss.item()
        predictions = torch.argmax(outputs, dim=1)

        train_preds.extend(predictions.detach().cpu().numpy())

        train_targets.extend(labels.detach().cpu().numpy())

    avg_train_loss = total_loss / len(train_loader)
    train_acc = accuracy_score(train_targets, train_preds)

    train_losses.append(avg_train_loss)
    train_accuracies.append(train_acc)

    print(
        f"Train Loss: {avg_train_loss:.4f}"
    )
    print(
    f"Train Accuracy: {train_acc:.4f}"
    )

    # =====================================================
    # VALIDATION
    # =====================================================

    model.eval()

    preds = []

    targets = []

    val_loss = 0

    with torch.no_grad():

        for batch in val_loader:

            pixel_values = batch["pixel_values"].to(device)

            input_ids = batch["input_ids"].to(device)

            attention_mask = batch["attention_mask"].to(device)

            labels = batch["label_B"].to(device)

            outputs = model(

                pixel_values,

                input_ids,

                attention_mask
            )

            loss = criterion(
                outputs,
                labels
            )

            val_loss += loss.item()

            predictions = torch.argmax(
                outputs,
                dim=1
            )

            preds.extend(
                predictions.cpu().numpy()
            )

            targets.extend(
                labels.cpu().numpy()
            )

    avg_val_loss = val_loss / len(val_loader)

    acc = accuracy_score(
        targets,
        preds
    )

    f1 = f1_score(
        targets,
        preds,
        average="weighted"
    )

    val_losses.append(avg_val_loss)

    val_accuracies.append(acc)

    val_f1s.append(f1)

    print("Validation Loss:", avg_val_loss)

    print("Validation Accuracy:", acc)

    print("Validation F1:", f1)

    # =====================================================
    # SAVE BEST MODEL
    # =====================================================

    if acc > best_score:

        best_score = acc

        torch.save({

            'epoch': epoch,

            'model_state_dict': model.state_dict(),

            'optimizer_state_dict': optimizer.state_dict(),

            'best_score': best_score

        }, "/content/drive/MyDrive/best_finetuned_clip_frequency_LAST.pt")

        print("💾 Best model saved")

# =========================================================
# FINAL TESTING
# =========================================================

print("\n🚀 FINAL TESTING")

checkpoint = torch.load(
    "/content/drive/MyDrive/best_finetuned_clip_frequency_LAST.pt",
    map_location=device
)

model.load_state_dict(
    checkpoint['model_state_dict']
)

optimizer.load_state_dict(
    checkpoint['optimizer_state_dict']
)

model.eval()

preds = []

targets = []

all_probs = []

test_loss = 0

with torch.no_grad():

    for batch in tqdm(test_loader):

        pixel_values = batch["pixel_values"].to(device)

        input_ids = batch["input_ids"].to(device)

        attention_mask = batch["attention_mask"].to(device)

        labels = batch["label_B"].to(device)

        outputs = model(

            pixel_values,

            input_ids,

            attention_mask
        )

        probabilities = torch.softmax(
            outputs,
            dim=1
        )

        all_probs.extend(
            probabilities.cpu().numpy()
        )

        loss = criterion(
            outputs,
            labels
        )

        test_loss += loss.item()

        predictions = torch.argmax(
            outputs,
            dim=1
        )

        preds.extend(
            predictions.cpu().numpy()
        )

        targets.extend(
            labels.cpu().numpy()
        )

# =========================================================
# TRAINING VS VALIDATION ACCURACY
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(train_accuracies)+1),
    train_accuracies,
    marker='o',
    linewidth=2,
    label='Training Accuracy'
)

plt.plot(
    range(1, len(val_accuracies)+1),
    val_accuracies,
    marker='o',
    linewidth=2,
    label='Validation Accuracy'
)

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# =========================================================
# TASK B RESULTS
# =========================================================
avg_test_loss = test_loss / len(test_loader)

acc = accuracy_score(targets, preds)

f1 = f1_score(
    targets,
    preds,
    average="weighted"
)

print("\nFINAL TASK B RESULTS")

print("Test Loss:", avg_test_loss)

print("Task B Accuracy:", acc)

print("Task B F1 Score:", f1)

print(
    classification_report(
        targets,
        preds
    )
)

# =========================================================
# TASK B CONFUSION MATRIX
# =========================================================
cm = confusion_matrix(
    targets,
    preds
)

plt.figure(figsize=(10,8))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("True")

plt.title("Task B Confusion Matrix")

plt.show()

# =========================================================
# TRAIN VS VALIDATION LOSS
# =========================================================
plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(train_losses)+1),
    train_losses,
    marker='o',
    linewidth=2,
    label='Train Loss'
)

plt.plot(
    range(1, len(val_losses)+1),
    val_losses,
    marker='o',
    linewidth=2,
    label='Validation Loss'
)

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("Training vs Validation Loss")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# VALIDATION ACCURACY GRAPH
# =========================================================
plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(val_accuracies)+1),
    val_accuracies,
    marker='o',
    linewidth=2,
    label='Validation Accuracy'
)

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.title("Validation Accuracy")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# VALIDATION F1 GRAPH
# =========================================================
plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(val_f1s)+1),
    val_f1s,
    marker='o',
    linewidth=2,
    label='Validation F1'
)

plt.xlabel("Epoch")

plt.ylabel("F1 Score")

plt.title("Validation F1 Score")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK B PRECISION / RECALL / F1 GRAPH
# =========================================================
task_b_report = classification_report(
    targets,
    preds,
    output_dict=True
)

task_b_classes = ['0', '1', '2', '3', '4', '5']

task_b_precision = [
    task_b_report[c]["precision"]
    for c in task_b_classes
]

task_b_recall = [
    task_b_report[c]["recall"]
    for c in task_b_classes
]

task_b_f1_values = [
    task_b_report[c]["f1-score"]
    for c in task_b_classes
]

x = np.arange(len(task_b_classes))

width = 0.25

plt.figure(figsize=(12,6))

plt.bar(
    x - width,
    task_b_precision,
    width,
    label='Precision'
)

plt.bar(
    x,
    task_b_recall,
    width,
    label='Recall'
)

plt.bar(
    x + width,
    task_b_f1_values,
    width,
    label='F1'
)

plt.xticks(x, task_b_classes)

plt.xlabel("Class")

plt.ylabel("Score")

plt.title("Task B Precision / Recall / F1")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK B ROC CURVES
# =========================================================
targets_bin = label_binarize(
    targets,
    classes=[0,1,2,3,4,5]
)

all_probs = np.array(all_probs)

fpr = dict()

tpr = dict()

roc_auc = dict()

for i in range(6):

    fpr[i], tpr[i], _ = roc_curve(
        targets_bin[:, i],
        all_probs[:, i]
    )

    roc_auc[i] = auc(
        fpr[i],
        tpr[i]
    )

plt.figure(figsize=(10,8))

for i in range(6):

    plt.plot(
        fpr[i],
        tpr[i],
        linewidth=2,
        label=f'Class {i} AUC = {roc_auc[i]:.3f}'
    )

plt.plot(
    [0,1],
    [0,1],
    linestyle='--'
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("Task B ROC Curves")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK B PR CURVES
# =========================================================
precision_dict = dict()

recall_dict = dict()

ap_dict = dict()

for i in range(6):

    precision_dict[i], recall_dict[i], _ = precision_recall_curve(

        targets_bin[:, i],

        all_probs[:, i]
    )

    ap_dict[i] = average_precision_score(

        targets_bin[:, i],

        all_probs[:, i]
    )

plt.figure(figsize=(10,8))

for i in range(6):

    plt.plot(

        recall_dict[i],

        precision_dict[i],

        linewidth=2,

        label=f'Class {i} AP = {ap_dict[i]:.3f}'
    )

plt.xlabel("Recall")

plt.ylabel("Precision")

plt.title("Task B Precision-Recall Curves")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK A RESULTS
# =========================================================
binary_targets = [0 if x == 0 else 1 for x in targets]

binary_preds = [0 if x == 0 else 1 for x in preds]

task_a_acc = accuracy_score(
    binary_targets,
    binary_preds
)

task_a_f1 = f1_score(
    binary_targets,
    binary_preds,
    average="weighted"
)

print("\n============================")

print("TASK A RESULTS")

print("============================")

print("Task A Accuracy:", task_a_acc)

print("Task A F1 Score:", task_a_f1)

print(
    classification_report(
        binary_targets,
        binary_preds,
        target_names=["Real", "Fake"]
    )
)

# =========================================================
# TASK A CONFUSION MATRIX
# =========================================================
task_a_cm = confusion_matrix(
    binary_targets,
    binary_preds
)

plt.figure(figsize=(6,5))

sns.heatmap(
    task_a_cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("True")

plt.title("Task A Confusion Matrix")

plt.show()

# =========================================================
# TASK A PRECISION / RECALL / F1 GRAPH
# =========================================================
task_a_report = classification_report(
    binary_targets,
    binary_preds,
    target_names=["Real", "Fake"],
    output_dict=True
)

task_a_classes = ["Real", "Fake"]

task_a_precision = [
    task_a_report[c]["precision"]
    for c in task_a_classes
]

task_a_recall = [
    task_a_report[c]["recall"]
    for c in task_a_classes
]

task_a_f1_values = [
    task_a_report[c]["f1-score"]
    for c in task_a_classes
]

x = np.arange(len(task_a_classes))

width = 0.25

plt.figure(figsize=(8,5))

plt.bar(
    x - width,
    task_a_precision,
    width,
    label='Precision'
)

plt.bar(
    x,
    task_a_recall,
    width,
    label='Recall'
)

plt.bar(
    x + width,
    task_a_f1_values,
    width,
    label='F1'
)

plt.xticks(x, task_a_classes)

plt.xlabel("Class")

plt.ylabel("Score")

plt.title("Task A Precision / Recall / F1")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK A ROC CURVE
# =========================================================
binary_probs = 1 - all_probs[:,0]

fpr, tpr, _ = roc_curve(
    binary_targets,
    binary_probs
)

roc_auc = auc(
    fpr,
    tpr
)

plt.figure(figsize=(7,6))

plt.plot(
    fpr,
    tpr,
    linewidth=2,
    label=f'AUC = {roc_auc:.3f}'
)

plt.plot(
    [0,1],
    [0,1],
    linestyle='--'
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("Task A ROC Curve")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK A PR CURVE
# =========================================================
precision_curve, recall_curve, _ = precision_recall_curve(

    binary_targets,

    binary_probs
)

ap_score = average_precision_score(

    binary_targets,

    binary_probs
)

plt.figure(figsize=(7,6))

plt.plot(

    recall_curve,

    precision_curve,

    linewidth=2,

    label=f'AP = {ap_score:.3f}'
)

plt.xlabel("Recall")

plt.ylabel("Precision")

plt.title("Task A Precision-Recall Curve")

plt.legend()

plt.grid(True)

plt.show()
```

```python
# =========================================================
# VISUALIZATION PIPELINE
# EXACTLY MATCHES TRAINING PREPROCESSING
# NO AVERAGE AUTOCORRELATION
# =========================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from datasets import load_dataset

# =========================================================
# LOAD DATASET
# =========================================================

dataset = load_dataset(
    "Rajarshi-Roy-research/Defactify_Image_Dataset"
)

test_ds = dataset["test"]

# =========================================================
# SETTINGS
# =========================================================

IMAGE_SIZE = 256

IMAGE_KEY = "Image"

# =========================================================
# SELECT SAMPLE IMAGE
# =========================================================

sample = test_ds[0]

pil_image = sample[IMAGE_KEY]

# =========================================================
# RGB → NUMPY
# =========================================================

image = np.array(pil_image)

# =========================================================
# RGB → GRAYSCALE
# =========================================================

gray = cv2.cvtColor(
    image,
    cv2.COLOR_RGB2GRAY
).astype(np.float32) / 255.0

# =========================================================
# WEAK DENOISING
# =========================================================

denoised = cv2.GaussianBlur(
    gray,
    (3, 3),
    0
)

# =========================================================
# RESIDUAL EXTRACTION
# r = x - D(x)
# =========================================================

residual = gray - denoised


# =========================================================
# FFT
# =========================================================

fft = np.fft.fft2(
    residual
)

fft_shift = np.fft.fftshift(
    fft
)

# =========================================================
# POWER SPECTRUM
# =========================================================

power = np.abs(
    fft_shift
) ** 2

# =========================================================
# AUTOCORRELATION
# =========================================================

autocorr = np.fft.ifft2(
    power
)

autocorr = np.real(
    autocorr
)

autocorr = np.fft.fftshift(
    autocorr
)
# =========================================================
# LOG SCALE
# =========================================================

autocorr = np.log1p(
    np.abs(autocorr)
)


# =========================================================
# NORMALIZATION
# =========================================================

autocorr = (
    autocorr - autocorr.min()
) / (
    autocorr.max() - autocorr.min() + 1e-8
)


# =========================================================
# CONVERT TO UINT8
# =========================================================

autocorr_uint8 = (
    autocorr * 255
).astype(np.uint8)

# =========================================================
# GRAYSCALE → RGB
# =========================================================

autocorr_rgb = cv2.cvtColor(
    autocorr_uint8,
    cv2.COLOR_GRAY2RGB
)

# =========================================================
# PIL IMAGE
# =========================================================

final_output = Image.fromarray(
    autocorr_rgb
)

# =========================================================
# DISPLAY RESULTS
# =========================================================

fig, axes = plt.subplots(
    2,
    4,
    figsize=(22, 10)
)

# =====================================================
# ORIGINAL IMAGE
# =====================================================

axes[0,0].imshow(image)

axes[0,0].set_title(
    "1. Original RGB Image",
    fontsize=14
)

axes[0,0].axis("off")

# =====================================================
# GRAYSCALE
# =====================================================

axes[0,1].imshow(
    gray,
    cmap="gray"
)

axes[0,1].set_title(
    "2. Grayscale",
    fontsize=14
)

axes[0,1].axis("off")

# =====================================================
# DENOISED
# =====================================================

axes[0,2].imshow(
    denoised,
    cmap="gray"
)

axes[0,2].set_title(
    "3. Denoised Image",
    fontsize=14
)

axes[0,2].axis("off")

# =====================================================
# RESIDUAL
# =====================================================

axes[0,3].imshow(
    residual,
    cmap="gray"
)

axes[0,3].set_title(
    "4. Residual Image",
    fontsize=14
)

axes[0,3].axis("off")

# =====================================================
# FFT SPECTRUM
# =====================================================

axes[1,0].imshow(
    power,
    cmap="viridis"
)

axes[1,0].set_title(
    "5. FFT Spectrum",
    fontsize=14
)

axes[1,0].axis("off")

# =====================================================
# AUTOCORRELATION
# =====================================================

axes[1,1].imshow(
    autocorr,
    cmap="viridis"
)

axes[1,1].set_title(
    "6. Normalized Autocorrelation",
    fontsize=14
)

axes[1,1].axis("off")

# =====================================================
# UINT8 AUTOCORRELATION
# =====================================================

axes[1,2].imshow(
    autocorr_uint8,
    cmap="gray"
)

axes[1,2].set_title(
    "7. Uint8 Autocorrelation",
    fontsize=14
)

axes[1,2].axis("off")

# =====================================================
# RGB OUTPUT
# =====================================================

axes[1,3].imshow(
    autocorr_rgb
)

axes[1,3].set_title(
    "8. Final RGB Output",
    fontsize=14
)

axes[1,3].axis("off")

plt.tight_layout()

plt.show()
```
