```python
# =========================================================
# IMPORTS
# =========================================================
import os
import random
import numpy as np

import torch
import torch.nn as nn

from torch.utils.data import Dataset, DataLoader

from datasets import load_dataset

from transformers import (
    CLIPProcessor,
    CLIPModel
)

from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns

from torchvision import transforms

from sklearn.preprocessing import label_binarize

from sklearn.metrics import (

    accuracy_score,

    f1_score,

    classification_report,

    precision_score,

    recall_score,

    roc_curve,

    auc,

    precision_recall_curve,

    average_precision_score
)

from tqdm import tqdm

from google.colab import drive

drive.mount('/content/drive')

# =========================================================
# GLOBAL SEED
# =========================================================
SEED = 42

random.seed(SEED)

np.random.seed(SEED)

torch.manual_seed(SEED)

torch.cuda.manual_seed(SEED)

torch.cuda.manual_seed_all(SEED)

os.environ["PYTHONHASHSEED"] = str(SEED)

torch.backends.cudnn.deterministic = True

torch.backends.cudnn.benchmark = False

torch.use_deterministic_algorithms(True)

# =========================================================
# DEVICE
# =========================================================
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("🚀 Device:", device)

# =========================================================
# LOAD DATASET
# =========================================================
dataset = load_dataset(
    "Rajarshi-Roy-research/Defactify_Image_Dataset"
)

train_ds = dataset["train"]

val_ds   = dataset["validation"]

test_ds  = dataset["test"]

# =========================================================
# PROCESSOR
# =========================================================
processor = CLIPProcessor.from_pretrained(
    "openai/clip-vit-base-patch32"
)

# =========================================================
# AUGMENTATION
# =========================================================
train_transform = transforms.Compose([

    transforms.RandomResizedCrop(
        224,
        scale=(0.75, 1.0)
    ),

    transforms.RandomHorizontalFlip(),

    transforms.RandomRotation(10),

    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2,
        saturation=0.2
    ),

    transforms.RandomGrayscale(p=0.05),
])

# =========================================================
# DATASET
# =========================================================
class ImageOnlyDataset(Dataset):

    def __init__(self, hf_dataset, train=False):

        self.dataset = hf_dataset

        self.train = train

    def __len__(self):

        return len(self.dataset)

    def __getitem__(self, idx):

        item = self.dataset[idx]

        image = item["Image"].convert("RGB")

        if self.train:
            image = train_transform(image)

        encoding = processor(

            images=image,

            return_tensors="pt"
        )

        return {

            "pixel_values":
                encoding["pixel_values"].squeeze(0),

            "label_B":
                torch.tensor(
                    item["Label_B"],
                    dtype=torch.long
                )
        }

# =========================================================
# DATALOADER SEEDING
# =========================================================
def seed_worker(worker_id):

    worker_seed = SEED + worker_id

    np.random.seed(worker_seed)

    random.seed(worker_seed)

g = torch.Generator()

g.manual_seed(SEED)

# =========================================================
# DATALOADERS
# =========================================================
BATCH_SIZE = 32

train_loader = DataLoader(
    ImageOnlyDataset(train_ds, train=True),
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

val_loader = DataLoader(
    ImageOnlyDataset(val_ds),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

test_loader = DataLoader(
    ImageOnlyDataset(test_ds),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    worker_init_fn=seed_worker,
    generator=g
)

# =========================================================
# MODEL
# =========================================================
class CleanerImageOnlyCLIP(nn.Module):

    def __init__(self):

        super().__init__()

        self.clip = CLIPModel.from_pretrained(
            "openai/clip-vit-base-patch32"
        )

        hidden_dim = self.clip.config.projection_dim
        # =================================================
        # FREEZE CLIP
        # =================================================
        for param in self.clip.parameters():
            param.requires_grad = False
        # =================================================
        # SAME FUSION NETWORK
        # =================================================
        self.fusion = nn.Sequential(

            nn.Linear(hidden_dim, 256),

            nn.LayerNorm(256),

            nn.GELU(),

            nn.Dropout(0.5),

            nn.Linear(256, 128),

            nn.LayerNorm(128),

            nn.GELU(),

            nn.Dropout(0.5)
        )
        # =================================================
        # SAME 6-CLASS HEAD
        # =================================================
        self.classifier = nn.Linear(128, 6)

    def forward(self, pixel_values):

        with torch.no_grad():

            outputs = self.clip.vision_model(
                pixel_values=pixel_values
            )

            pooled_output = outputs.pooler_output

            image_feat = self.clip.visual_projection(
                pooled_output
            )

        fused = self.fusion(image_feat)

        logits = self.classifier(fused)

        return logits

# =========================================================
# MODEL
# =========================================================
model = CleanerImageOnlyCLIP().to(device)

# =========================================================
# LOSS
# =========================================================
criterion = nn.CrossEntropyLoss(
    label_smoothing=0.1
)

# =========================================================
# OPTIMIZER
# =========================================================
optimizer = torch.optim.AdamW(

    model.parameters(),

    lr=5e-6,

    weight_decay=1e-3
)

# =========================================================
# HISTORY
# =========================================================
train_losses = []

val_losses = []

val_accuracies = []

val_f1s = []

# =========================================================
# TRAINING
# =========================================================
best_score = 0

EPOCHS = 25

for epoch in range(EPOCHS):

    print(f"\n🔥 Epoch {epoch+1}/{EPOCHS}")

    model.train()

    total_loss = 0

    for batch in tqdm(train_loader):

        pixel_values = batch["pixel_values"].to(device)

        labels = batch["label_B"].to(device)

        optimizer.zero_grad()

        outputs = model(pixel_values)

        loss = criterion(outputs, labels)

        loss.backward()

        torch.nn.utils.clip_grad_norm_(
            model.parameters(),
            1.0
        )

        optimizer.step()

        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_loader)

    train_losses.append(avg_train_loss)

    print(f"Train Loss: {avg_train_loss:.4f}")

    # =====================================================
    # VALIDATION
    # =====================================================
    model.eval()

    val_loss = 0

    preds = []

    targets = []

    with torch.no_grad():

        for batch in val_loader:

            pixel_values = batch["pixel_values"].to(device)

            labels = batch["label_B"].to(device)

            outputs = model(pixel_values)

            loss = criterion(outputs, labels)

            val_loss += loss.item()

            predictions = torch.argmax(outputs, dim=1)

            preds.extend(predictions.cpu().numpy())

            targets.extend(labels.cpu().numpy())

    avg_val_loss = val_loss / len(val_loader)

    acc = accuracy_score(targets, preds)

    f1 = f1_score(
        targets,
        preds,
        average="weighted"
    )

    val_losses.append(avg_val_loss)

    val_accuracies.append(acc)

    val_f1s.append(f1)

    print("Validation Loss:", avg_val_loss)

    print("Validation Accuracy:", acc)

    print("Validation F1:", f1)

    # =====================================================
    # SAVE BEST MODEL
    # =====================================================
    if acc > best_score:

        best_score = acc

        torch.save({

            'epoch': epoch,

            'model_state_dict': model.state_dict(),

            'optimizer_state_dict': optimizer.state_dict(),

            'best_score': best_score

        }, "/content/drive/MyDrive/best_image_only_clip_6class.pt")

        print("💾 Best model saved")

# =========================================================
# FINAL TEST
# =========================================================

print("\n🚀 FINAL TESTING")

checkpoint = torch.load(
    "/content/drive/MyDrive/best_image_only_clip_6class.pt",
    map_location=device
)

model.load_state_dict(
    checkpoint['model_state_dict']
)

optimizer.load_state_dict(
    checkpoint['optimizer_state_dict']
)

print("✅ Best checkpoint loaded")

model.eval()

test_loss = 0

preds = []

targets = []

all_probs = []

with torch.no_grad():

    for batch in tqdm(test_loader):

        pixel_values = batch["pixel_values"].to(device)

        labels = batch["label_B"].to(device)

        outputs = model(pixel_values)

        probabilities = torch.softmax(outputs, dim=1)

        all_probs.extend(
            probabilities.cpu().numpy()
        )

        loss = criterion(outputs, labels)

        test_loss += loss.item()

        predictions = torch.argmax(outputs, dim=1)

        preds.extend(predictions.cpu().numpy())

        targets.extend(labels.cpu().numpy())

# =========================================================
# TASK B RESULTS
# =========================================================

avg_test_loss = test_loss / len(test_loader)

acc = accuracy_score(targets, preds)

f1 = f1_score(
    targets,
    preds,
    average="weighted"
)

print("\nFINAL TASK B RESULTS")

print("Test Loss:", avg_test_loss)

print("Task B Accuracy:", acc)

print("Task B F1 Score:", f1)

print(
    classification_report(
        targets,
        preds
    )
)

# =========================================================
# TASK B ROC CURVES
# =========================================================

targets_bin = label_binarize(
    targets,
    classes=[0,1,2,3,4,5]
)

all_probs = np.array(all_probs)

fpr = dict()

tpr = dict()

roc_auc = dict()

for i in range(6):

    fpr[i], tpr[i], _ = roc_curve(
        targets_bin[:, i],
        all_probs[:, i]
    )

    roc_auc[i] = auc(
        fpr[i],
        tpr[i]
    )

plt.figure(figsize=(10,8))

for i in range(6):

    plt.plot(
        fpr[i],
        tpr[i],
        linewidth=2,
        label=f'Class {i} AUC = {roc_auc[i]:.3f}'
    )

plt.plot([0,1],[0,1], linestyle='--')

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("Task B ROC Curves")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK B PR CURVES
# =========================================================

precision_dict = dict()

recall_dict = dict()

ap_dict = dict()

for i in range(6):

    precision_dict[i], recall_dict[i], _ = precision_recall_curve(

        targets_bin[:, i],

        all_probs[:, i]
    )

    ap_dict[i] = average_precision_score(

        targets_bin[:, i],

        all_probs[:, i]
    )

plt.figure(figsize=(10,8))

for i in range(6):

    plt.plot(

        recall_dict[i],

        precision_dict[i],

        linewidth=2,

        label=f'Class {i} AP = {ap_dict[i]:.3f}'
    )

plt.xlabel("Recall")

plt.ylabel("Precision")

plt.title("Task B Precision-Recall Curves")

plt.legend()

plt.grid(True)

plt.show()
# =========================================================
# TASK B CONFUSION MATRIX
# =========================================================

cm = confusion_matrix(
    targets,
    preds
)

plt.figure(figsize=(10,8))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("True")

plt.title("Task B Confusion Matrix")

plt.show()

# =========================================================
# TRAIN VS VALIDATION LOSS
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(train_losses)+1),
    train_losses,
    marker='o',
    linewidth=2,
    label='Train Loss'
)

plt.plot(
    range(1, len(val_losses)+1),
    val_losses,
    marker='o',
    linewidth=2,
    label='Validation Loss'
)

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("Training vs Validation Loss")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# VALIDATION ACCURACY GRAPH
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(val_accuracies)+1),
    val_accuracies,
    marker='o',
    linewidth=2,
    label='Validation Accuracy'
)

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.title("Validation Accuracy")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# VALIDATION F1 GRAPH
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(val_f1s)+1),
    val_f1s,
    marker='o',
    linewidth=2,
    label='Validation F1'
)

plt.xlabel("Epoch")

plt.ylabel("F1 Score")

plt.title("Validation F1 Score")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK A RESULTS
# =========================================================

binary_targets = [0 if x == 0 else 1 for x in targets]

binary_preds = [0 if x == 0 else 1 for x in preds]

task_a_acc = accuracy_score(
    binary_targets,
    binary_preds
)

task_a_f1 = f1_score(
    binary_targets,
    binary_preds,
    average="weighted"
)

print("\n============================")

print("TASK A RESULTS")

print("============================")

print("Task A Accuracy:", task_a_acc)

print("Task A  F1 Score:", task_a_f1)

print(
    classification_report(
        binary_targets,
        binary_preds,
        target_names=["Real", "Fake"]
    )
)

# =========================================================
# TASK A CONFUSION MATRIX
# =========================================================

task_a_cm = confusion_matrix(
    binary_targets,
    binary_preds
)

plt.figure(figsize=(6,5))

sns.heatmap(
    task_a_cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("True")

plt.title("Task A Confusion Matrix")

plt.show()

# =========================================================
# TASK A ROC CURVE
# =========================================================

binary_probs = 1 - all_probs[:,0]

fpr, tpr, _ = roc_curve(
    binary_targets,
    binary_probs
)

roc_auc = auc(
    fpr,
    tpr
)

plt.figure(figsize=(7,6))

plt.plot(
    fpr,
    tpr,
    linewidth=2,
    label=f'AUC = {roc_auc:.3f}'
)

plt.plot(
    [0,1],
    [0,1],
    linestyle='--'
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("Task A ROC Curve")

plt.legend()

plt.grid(True)

plt.show()

# =========================================================
# TASK A PRECISION-RECALL CURVE
# =========================================================

precision_curve, recall_curve, _ = precision_recall_curve(

    binary_targets,

    binary_probs
)

ap_score = average_precision_score(

    binary_targets,

    binary_probs
)

plt.figure(figsize=(7,6))

plt.plot(

    recall_curve,

    precision_curve,

    linewidth=2,

    label=f'AP = {ap_score:.3f}'
)

plt.xlabel("Recall")

plt.ylabel("Precision")

plt.title("Task A Precision-Recall Curve")

plt.legend()

plt.grid(True)

plt.show()
```
